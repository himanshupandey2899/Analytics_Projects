## Part 2: Customer Analysis

This section analyzes customer demographics, income levels, and spending behaviour to identify meaningful business insights that support customer segmentation and marketing decisions.

### 1. What is the average age of customers?

The average customer age provides insight into the dominant demographic group visiting the shopping mall.

```python
df['age'].mean().__round__()
```

**Insight:**  
The average customer age indicates the primary age demographic of mall visitors, helping businesses design age-appropriate marketing campaigns and product offerings.

### 2. What is the average annual income of customers?

Understanding customer income levels helps assess purchasing power and pricing strategy suitability.

```python
df['annual_income'].mean().__round__(2)
```

**Insight:**  
The average annual income reflects the overall economic profile of customers and helps businesses align pricing strategies with customer affordability.

### 3. Which gender spends more on average?

Analyzing spending behaviour across gender groups helps businesses design targeted promotions and personalized marketing strategies.

```python
df.groupby('gender')['spending_score'].mean().round(2)
```

```python
df.groupby('gender')['spending_score'].mean().plot(kind='bar')
plt.title("Average Spending Score by Gender")
plt.ylabel("Average Spending Score")
plt.show()
```

**Insight:**  
Comparing average spending scores across genders highlights purchasing behaviour differences and enables gender-focused marketing initiatives.

### 4. Which age group has the highest spending score?

Customers are grouped into age categories to identify the most valuable spending demographic.

```python
df['age_group'] = pd.cut(
    df['age'],
    bins=[18,25,35,45,55,65,75,90],
    labels=['18-25','26-35','36-45','46-55','56-65','66-75','76-90']
)
```

```python
df.groupby('age_group')['spending_score'].mean().round(2)
```

```python
df.groupby('age_group')['spending_score'].mean().plot(kind='bar')
plt.title("Average Spending Score by Age Group")
plt.ylabel("Average Spending Score")
plt.show()
```

**Insight:**  
Identifying the highest-spending age group enables businesses to prioritize marketing investments toward the most revenue-generating customer segment.

### 5. Is there a relationship between annual income and spending score?

Analyzing the relationship between income and spending behaviour helps determine whether higher income customers necessarily spend more.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='annual_income', y='spending_score')
plt.title("Annual Income vs Spending Score")
plt.show()
```

```python
df[['annual_income','spending_score']].corr()
```

**Insight:**  
The relationship between income and spending score reveals whether purchasing behaviour depends primarily on income level or other behavioral factors.
