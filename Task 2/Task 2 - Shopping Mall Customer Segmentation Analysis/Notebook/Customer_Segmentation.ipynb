## Part 3: Customer Segmentation

Customer segmentation groups customers based on income levels and spending behaviour. 

Segmentation enables businesses to:
- identify high-value customers,
- personalize marketing strategies,
- improve customer retention,
- optimize revenue generation.

### Step 1: Define Income and Spending Thresholds

Customers are categorized as High or Low based on the average annual income and spending score values.

| Income | Spending | Segment             |
| ------ | -------- | ------------------- |
| High   | High     | Premium Customers   |
| High   | Low      | Potential Customers |
| Low    | High     | Loyal Customers     |
| Low    | Low      | Budget Customers    |

```python
# Define thresholds for segmentation

income_threshold = df['annual_income'].mean().round(2)
spending_threshold = df['spending_score'].mean().round(2)

income_threshold, spending_threshold
```

### Step 2: Create Customer Segments

Customers are assigned to segments based on income level and spending behaviour.

```python
# Define segmentation function

def segment_customer(row):
    if row['annual_income'] >= income_threshold and row['spending_score'] >= spending_threshold:
        return 'High Income - High Spending'
    elif row['annual_income'] >= income_threshold and row['spending_score'] < spending_threshold:
        return 'High Income - Low Spending'
    elif row['annual_income'] < income_threshold and row['spending_score'] >= spending_threshold:
        return 'Low Income - High Spending'
    else:
        return 'Low Income - Low Spending'

df['segment'] = df.apply(segment_customer, axis=1)

df.head(3)
```

### Step 3: Segment Distribution Analysis

The number of customers in each segment is analyzed to understand customer composition.

```python
# Analyze Segment Distribution

segment_counts = df['segment'].value_counts()
segment_counts
```

```python
# Visualize Segment Distribution

segment_counts.plot(kind='bar')
plt.title("Customer Segment Distribution")
plt.xlabel("Segment")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.show()
```

### Step 4: Segment Characteristics

Average income and spending behaviour are examined for each customer segment.

```python
df.groupby('segment')[['annual_income','spending_score','age']].mean().round()
```

### Step 5 — Segmentation Visualization

### Income vs Spending Segmentation Visualization

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='annual_income',
    y='spending_score',
    hue='segment'
)

plt.title("Customer Segmentation Based on Income and Spending")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

### Advanced Segmentation using K-Means Clustering

K-Means clustering is applied to identify natural customer groupings based on behavioural similarity.

```python
from sklearn.cluster import KMeans

X = df[['annual_income','spending_score']]

kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x=df['annual_income'],
    y=df['spending_score'],
    hue=df['cluster'],
    palette='Set2'
)

plt.title("K-Means Customer Clusters")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```
